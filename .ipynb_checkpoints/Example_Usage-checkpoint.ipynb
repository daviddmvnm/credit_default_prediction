### Preparing Data with unified pipeline and loading the model

```python
from pipelines import unified_pipeline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from joblib import load

#1 Import the unproccessed datasets to use pipelines on

ap_test_df = pd.read_csv("application_test.csv")
cr_df = pd.read_csv("credit_card_balance.csv")

#2 Run unified pipeline on `application_test` and "credit_card_balance" datasets
model_ready_data = unified_pipeline(ap_test_df, cr_df, include_target=False)
print("data prepared succesfully")

#3 load the model
model = load("log_reg_oversample_model.joblib")  
print("Model loaded successfully!")
```

# Example Usage
## **Categorising Based on Predicted Risk**

```python
#Making predictions with the model
predictions = model.predict(model_ready_data)
prediction_probabilities = model.predict_proba(model_ready_data)[:, 1]  # Probability of class 1 (default)

#Display predictions with `SK_ID_CURR` as the index
predictions_df = pd.DataFrame({
    "Predicted Default": predictions,
    "Default Probability": prediction_probabilities
}, index=model_ready_data.index) 

print("Model predictions:")
print(predictions_df.head())


# Define risk categories based on predicted probabilities
def categorize_risk(prob):
    if prob >= 0.8:
        return "Very High Risk"
    elif prob >= 0.7:
        return "High Risk"
    elif prob >= 0.4:
        return "Moderate Risk"
    else:
        return "Low Risk"

# Apply categorization
predictions_df["Risk Category"] = predictions_df["Default Probability"].apply(categorize_risk)

# Show a few results
predictions_df.head(10)
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of risk categories
risk_counts = predictions_df["Risk Category"].value_counts()
plt.figure(figsize=(8, 6))
risk_counts.plot(kind="bar")
plt.title("Distribution of Risk Categories")
plt.xlabel("Risk Category")
plt.ylabel("Number of Applicants")
plt.show()

# Display top n high-risk applicants for immeidate action
top_risk_applicants = predictions_df.sort_values(by="Default Probability", ascending=False).head(10)
top_risk_applicants

```

## Business Implications and Model Utility

This model provides default probabilities for each applicant, which can be categorized into actionable risk levels (e.g., High, Moderate, and Low Risk).

In a real-world application, a financial institution could use this model to:
- Focus review efforts on High-Risk applicants, potentially adding further verification steps.
- Offer better terms or faster processing to Low-Risk applicants, thus optimizing resources.
- Periodically retrain or update the model as more labeled data (defaults/non-defaults) becomes available.
